```python


import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.model_selection import train_test_split

# Define criteria for single and multiple bookings
booking_counts = df_clean.groupby(['Location', 'Class', 'Date', 'Code']).size()
single_booking_criteria = booking_counts == 1

# Filter data into single and multiple bookings datasets
single_booking_data = df_clean[df_clean.set_index(['Location', 'Class', 'Date', 'Code']).index.isin(booking_counts[single_booking_criteria].index)]
multiple_booking_data = df_clean[df_clean.set_index(['Location', 'Class', 'Date', 'Code']).index.isin(booking_counts[~single_booking_criteria].index)]

# Calculate count of bookings for single and multiple bookings
single_booking_data['count'] = single_booking_data.set_index(['Location', 'Class', 'Date', 'Code']).index.map(booking_counts)
multiple_booking_data['count'] = multiple_booking_data.set_index(['Location', 'Class', 'Date', 'Code']).index.map(booking_counts)

# Separate features and target variable for single bookings
X_single = single_booking_data.drop('count', axis=1)
y_single = single_booking_data['count']

# Separate features and target variable for multiple bookings
X_multiple = multiple_booking_data.drop('count', axis=1)
y_multiple = multiple_booking_data['count']

# One-hot encode categorical variables for single bookings
enc_single = OneHotEncoder()
X_single_enc = enc_single.fit_transform(X_single.select_dtypes(include=['object']))

# One-hot encode categorical variables for multiple bookings
enc_multiple = OneHotEncoder()
X_multiple_enc = enc_multiple.fit_transform(X_multiple.select_dtypes(include=['object']))


# Split the single booking data 
X_single_train, X_single_test, y_single_train, y_single_test = train_test_split(X_single_enc, y_single, test_size=0.2, random_state=42)

# Split the multiple booking data 
X_multiple_train, X_multiple_test, y_multiple_train, y_multiple_test = train_test_split(X_multiple_enc, y_multiple, test_size=0.2, random_state=42)

# Train linear regression model for single bookings
lreg_single = LinearRegression()
lreg_single.fit(X_single_train, y_single_train)
y_single_pred = lreg_single.predict(X_single_test)

# Train linear regression model for multiple bookings
lreg_multiple = LinearRegression()
lreg_multiple.fit(X_multiple_train, y_multiple_train)
y_multiple_pred = lreg_multiple.predict(X_multiple_test)

# Evaluate the model for single bookings
mse_single = mean_squared_error(y_single_test, y_single_pred)
print("MSE for single bookings:", mse_single)
rmse_single = np.sqrt(mse_single)
print("RMSE for single bookings:", rmse_single)
mae_single = mean_absolute_error(y_single_test, y_single_pred)
print("MAE for single bookings:", mae_single)
r2_single = r2_score(y_single_test, y_single_pred)
print("R-squared for single bookings:", r2_single)

# Evaluate the model for multiple bookings
mse_multiple = mean_squared_error(y_multiple_test, y_multiple_pred)
print("MSE for multiple bookings:", mse_multiple)
rmse_multiple = np.sqrt(mse_multiple)
print("RMSE for multiple bookings:", rmse_multiple)
mae_multiple = mean_absolute_error(y_multiple_test, y_multiple_pred)
print("MAE for multiple bookings:", mae_multiple)
r2_multiple = r2_score(y_multiple_test, y_multiple_pred)
print("R-squared for multiple bookings:", r2_multiple)


```

```python
from sklearn.ensemble import RandomForestRegressor

# Train Random Forest model for single bookings
rf_single = RandomForestRegressor()
rf_single.fit(X_single_train, y_single_train)
y_single_pred_rf = rf_single.predict(X_single_test)

# Train Random Forest model for multiple bookings
rf_multiple = RandomForestRegressor()
rf_multiple.fit(X_multiple_train, y_multiple_train)
y_multiple_pred_rf = rf_multiple.predict(X_multiple_test)

# Evaluate Random Forest model for single bookings
mse_single_rf = mean_squared_error(y_single_test, y_single_pred_rf)
print("MSE for single bookings (Random Forest):", mse_single_rf)
rmse_single_rf = np.sqrt(mse_single_rf)
print("RMSE for single bookings (Random Forest):", rmse_single_rf)
mae_single_rf = mean_absolute_error(y_single_test, y_single_pred_rf)
print("MAE for single bookings (Random Forest):", mae_single_rf)
r2_single_rf = r2_score(y_single_test, y_single_pred_rf)
print("R-squared for single bookings (Random Forest):", r2_single_rf)

# Evaluate Random Forest model for multiple bookings
mse_multiple_rf = mean_squared_error(y_multiple_test, y_multiple_pred_rf)
print("MSE for multiple bookings (Random Forest):", mse_multiple_rf)
rmse_multiple_rf = np.sqrt(mse_multiple_rf)
print("RMSE for multiple bookings (Random Forest):", rmse_multiple_rf)
mae_multiple_rf = mean_absolute_error(y_multiple_test, y_multiple_pred_rf)
print("MAE for multiple bookings (Random Forest):", mae_multiple_rf)
r2_multiple_rf = r2_score(y_multiple_test, y_multiple_pred_rf)
print("R-squared for multiple bookings (Random Forest):", r2_multiple_rf)
```

```python
from sklearn.ensemble import GradientBoostingRegressor

# Train Gradient Boosting model for single bookings
gb_single = GradientBoostingRegressor()
gb_single.fit(X_single_train, y_single_train)
y_single_pred_gb = gb_single.predict(X_single_test)

# Train Gradient Boosting model for multiple bookings
gb_multiple = GradientBoostingRegressor()
gb_multiple.fit(X_multiple_train, y_multiple_train)
y_multiple_pred_gb = gb_multiple.predict(X_multiple_test)

# Evaluate Gradient Boosting model for single bookings
mse_single_gb = mean_squared_error(y_single_test, y_single_pred_gb)
print("MSE for single bookings (Gradient Boosting):", mse_single_gb)
rmse_single_gb = np.sqrt(mse_single_gb)
print("RMSE for single bookings (Gradient Boosting):", rmse_single_gb)
mae_single_gb = mean_absolute_error(y_single_test, y_single_pred_gb)
print("MAE for single bookings (Gradient Boosting):", mae_single_gb)
r2_single_gb = r2_score(y_single_test, y_single_pred_gb)
print("R-squared for single bookings (Gradient Boosting):", r2_single_gb)

# Evaluate Gradient Boosting model for multiple bookings
mse_multiple_gb = mean_squared_error(y_multiple_test, y_multiple_pred_gb)
print("MSE for multiple bookings (Gradient Boosting):", mse_multiple_gb)
rmse_multiple_gb = np.sqrt(mse_multiple_gb)
print("RMSE for multiple bookings (Gradient Boosting):", rmse_multiple_gb)
mae_multiple_gb = mean_absolute_error(y_multiple_test, y_multiple_pred_gb)
print("MAE for multiple bookings (Gradient Boosting):", mae_multiple_gb)
r2_multiple_gb = r2_score(y_multiple_test, y_multiple_pred_gb)
print("R-squared for multiple bookings (Gradient Boosting):", r2_multiple_gb)
```
